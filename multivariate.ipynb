```python
# Load the Iris Dataset

from sklearn.datasets import load_iris
import pandas as pd

# Load the Iris dataset
iris = load_iris()
data = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Map species labels to their names
species_mapping = {0: 'setosa', 1: 'versicolor', 2: 'virginica'}
data['species'] = iris.target
data['species'] = data['species'].map(species_mapping)

data.head()
```

```python
# Create a pair plot for Multivariate Distribution
import seaborn as sns
import matplotlib.pyplot as plt

# Create pairplot for visualizing multivariate distribution
sns.pairplot(data, hue='species', palette='viridis')
plt.show()
```

```python
# Additional Visualization
# Box plot for each feature

plt.figure(figsize=(12, 6))
for i, feature in enumerate(iris.feature_names):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(x='species', y=feature, data=data)
    plt.title(f'Box plot of {feature}')
plt.tight_layout()
plt.show()

```

```python

# Violin plot for each feature

plt.figure(figsize=(12, 6))
for i, feature in enumerate(iris.feature_names):
    plt.subplot(2, 2, i + 1)
    sns.violinplot(x='species', y=feature, data=data, palette='viridis')
    plt.title(f'Violin plot of {feature}')
plt.tight_layout()
plt.show()
```
